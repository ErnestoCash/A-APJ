```python
# Step 1: Install CrewAI (uncomment the next line if you're running this in Colab)
!pip install crewai

import os

# Set your OpenAI API key here
os.environ['OPENAI_API_KEY'] = 'your_api_key_here'


# Step 2: Import necessary classes from CrewAI
from crewai import Agent, Task, Crew, Process

# Step 3: Define the agents
analyst = Agent(
    role='Senior social media analyst',
    goal='Research and analyze social media content for insights.',
    backstory="Expert in technology and startup trends, working for a personal branding agency.",
    verbose=True
)

content_creator = Agent(
    role='LinkedIn Content Creator Specialist',
    goal='Create engaging content for LinkedIn based on research insights.',
    backstory="Specializes in content creation for tech executives, focusing on AI and startups.",
    verbose=True
)

# Step 4: Define the tasks
task1 = Task(
    description='Identify trending topics in AI and startups for LinkedIn posts.',
    agent=analyst
)

task2 = Task(
    description='Develop a detailed LinkedIn post on a chosen trending topic.',
    agent=content_creator
)

# Step 5: Create and configure the crew
crew = Crew(
    agents=[analyst, content_creator],
    tasks=[task1, task2],
    process=Process.sequential,
    verbose=True
)

# Step 6: Execute the crew's workflow
result = crew.kickoff()

# Display the result
print(result)
```
